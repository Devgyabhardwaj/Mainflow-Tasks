```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\Devgya\\Desktop\\DATA ANAYLTYICS\\householdtask3.csv')
```

```python
data.head(10)
```

```python
#Sactter Plot against year to own
plt.scatter(data['year'],data['own'])
#adding titlt to the plot
plt.title("Scatter plot")
#Setting x&y lables
plt.xlabel('own')
plt.ylabel('year')
#showing the chart
plt.show()
```

```python
#Line chart against year to own
plt.plot(data['year'])
plt.plot(data['own'])
#adding titlt to the plot
plt.title("Line Chart")
#Setting x&y lables
plt.xlabel('own')
plt.ylabel('year')
#showing the chart
plt.show()
```

```python
# Bar chart against year to own
plt.bar(data['year'],data['own'])
#adding titlt to the plot
plt.title("Bar chart")
#Setting x&y lables
plt.xlabel('own')
plt.ylabel('year')
#showing the chart
plt.show()
```

```python
#plotting histogram for income
plt.hist(data['income'])

#giving title
plt.title("Histogram")

plt.show()
```
